## Construye el modelo

```python
def smape(y_true, y_pred):
    """
    Calcula el sMAPE entre dos vectores.
    """
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2
    diff = np.abs(y_true - y_pred)

    # Evitar divisi√≥n por cero
    mask = denominator != 0
    return np.mean((diff[mask] / denominator[mask])) * 100


def smape_final(rougher_true, rougher_pred, final_true, final_pred):
    """
    Calcula la m√©trica final del proyecto:
    sMAPE_final = 0.25 * sMAPE_rougher + 0.75 * sMAPE_final
    """

    smape_rougher = smape(rougher_true, rougher_pred)
    smape_final_val = smape(final_true, final_pred)

    return 0.25 * smape_rougher + 0.75 * smape_final_val
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import make_scorer
from sklearn.impute import SimpleImputer

# =====================================================
# 1. Funci√≥n sMAPE
# =====================================================
def smape(y_true, y_pred):
    return (100/len(y_true)) * np.sum(
        2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred))
    )

smape_scorer = make_scorer(smape, greater_is_better=False)

# =====================================================
# 2. QUEDARSE SOLO CON COLUMNAS NUM√âRICAS
# =====================================================
train_num = train.select_dtypes(include='number').copy()
test_num = test.select_dtypes(include='number').copy()

# =====================================================
# 3. REMOVER INF Y -INF
# =====================================================
train_num = train_num.replace([np.inf, -np.inf], np.nan)
test_num = test_num.replace([np.inf, -np.inf], np.nan)

# =====================================================
# 4. ELIMINAR FILAS DONDE LOS TARGETS NO EXISTEN
# =====================================================
train_num = train_num.dropna(subset=['rougher.output.recovery',
                                     'final.output.recovery'])

# =====================================================
# 5. SEPARAR TARGETS
# =====================================================
y_train_rougher = train_num['rougher.output.recovery']
y_train_final = train_num['final.output.recovery']

X_train = train_num.drop(['rougher.output.recovery', 'final.output.recovery'],
                         axis=1)

# =====================================================
# 6. ALINEAR COLUMNAS ENTRE TRAIN Y TEST
# =====================================================

# Crear columnas faltantes en test
for col in X_train.columns:
    if col not in test_num.columns:
        test_num[col] = np.nan

# Ordenar columnas igual que train
X_test = test_num[X_train.columns].copy()

# =====================================================
# 7. ELIMINAR FILAS QUE SEAN COMPLETAMENTE NULAS
# =====================================================
X_train = X_train.dropna(how='all')
X_test = X_test.dropna(how='all')

# Eliminar tambi√©n en los targets
idx_train_valid = X_train.index
y_train_rougher = y_train_rougher.loc[idx_train_valid]
y_train_final = y_train_final.loc[idx_train_valid]

# =====================================================
# 8. IMPUTACI√ìN ROBUSTA
# =====================================================
imputer = SimpleImputer(strategy='median')

X_train_clean = pd.DataFrame(imputer.fit_transform(X_train),
                             columns=X_train.columns)

X_test_clean = pd.DataFrame(imputer.transform(X_test),
                            columns=X_test.columns)

# Asegurar que NO haya NaN despu√©s de imputar
assert not X_train_clean.isnull().any().any(), "Quedan NaN en train"
assert not X_test_clean.isnull().any().any(), "Quedan NaN en test"


```

```python
# =====================================================
# 9. MODELADO
# =====================================================
models = {
    "LinearRegression": LinearRegression(),
    "RandomForest": RandomForestRegressor(n_estimators=200, random_state=42),
    "GradientBoosting": GradientBoostingRegressor(random_state=42),
}

results = {}

# =====================================================
# 10. VALIDACI√ìN CRUZADA
# =====================================================
for name, model in models.items():

    score_rougher = cross_val_score(
        model, X_train_clean, y_train_rougher, cv=5, scoring=smape_scorer
    ).mean()

    score_final = cross_val_score(
        model, X_train_clean, y_train_final, cv=5, scoring=smape_scorer
    ).mean()

    final_metric = 0.25 * score_rougher + 0.75 * score_final
    results[name] = final_metric

    print(f"{name} ‚Üí sMAPE Total = {final_metric}")

print("\nResultados finales:")
print(results)

best_model_name = min(results, key=results.get)
print("\nMejor modelo:", best_model_name)

```
